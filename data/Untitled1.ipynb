```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ["Oluwafemi", "Jane", "John", "Chris", np.nan, None, 'NA'],
    "last": ["Olasegiri", "Doe", "Doe", "Olasegiri", np.nan, np.nan, 'Missing'],
    "email": ["olasegirioluwafemi@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com", None, np.nan, "Anonymous@gmail.com", 'NA'],
    "age": ['33', '55', '63', '36', None, None, 'Missing'],
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.dropna()
```

```python
df.dropna(axis='index', how='any')
```

```python
df.dropna(axis='index', how='any', subset=['email'])
```

```python
df.dropna(axis='index', how='any', subset=['email', 'last', 'first'])
```

```python
df.replace('NA', np.nan, inplace=True)
```

```python
df.replace('missing', np.nan, inplace=True)
```
